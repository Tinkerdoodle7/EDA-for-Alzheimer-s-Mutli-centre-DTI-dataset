```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
data=loadmat('MCAD_AFQ_competition.mat', mat_dtype=True)
data
```

```python
data.keys()
```

```python
# Get the Keys and store them in a list
labels = list(data['train_population'])

# Get the[] Values and store them in a list
gender = [row[0] for row in labels]
```

```python
nn = [(data['train_sites']).tolist()[i][0] for i in range(len(data['train_sites']))]
```

```python
new_list = list(zip(gender, nn))
```

```python
df2 = pd.DataFrame(new_list, columns=['Gender', "Train_sites"])
```

```python
import seaborn as sns
```

```python
sns.catplot(y="Train_sites", hue="Gender", kind="count",
            palette="pastel", edgecolor=".6",
            data=df2)
```

```python
age = [data['train_population'][i][1] for i in range(len(data['train_population']))]
gender = [int(data['train_population'][i][0]) for i in range(len(data['train_population']))]
```

```python
new_list_2 = list(zip(nn, age))
df3 = pd.DataFrame(new_list_2, columns=['Train_sites', 'Age'])
```

```python
sns.swarmplot(x="Train_sites", y="Age", data=df3)
```
